```python
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
import pyecharts.options as opts
from pyecharts.charts import Line
from pyecharts.globals import ThemeType
```

```python
x_data = ["x1", "x2", "x3", "x4", "x5", "x6", "x7"]
y_data = [820, 932, 901, 934, 1290, 1330, 1320]

line = (
    Line(init_opts=opts.InitOpts(theme=ThemeType.DARK, width='1100px', height='550px'))
    .set_global_opts(
        title_opts=opts.TitleOpts(title="My Title"),
        legend_opts=opts.LegendOpts(pos_left="center", pos_top="7%"),
        xaxis_opts=opts.AxisOpts(
            name="X Axis", 
            splitarea_opts=opts.SplitLineOpts(is_show=True),
        ),
        tooltip_opts=opts.TooltipOpts(trigger="axis"),
        yaxis_opts=opts.AxisOpts(
            type_="value",
            name="main y axis",
            min_='dataMin',
            position="left",
            axistick_opts=opts.AxisTickOpts(is_show=True),
            splitline_opts=opts.SplitLineOpts(is_show=True),
            #is_scale=True,
        ),
        datazoom_opts=[
            opts.DataZoomOpts(range_start=0, range_end=100),
            opts.DataZoomOpts(
                type_="inside", range_start=0, range_end=100),
        ],)
    .add_xaxis(xaxis_data=x_data)
    .add_yaxis(
        series_name="yA",
        y_axis=[120, 132, 101, 134, 90, 230, 210],
        yaxis_index=0,
        label_opts=opts.LabelOpts(is_show=False),
    )
    .add_yaxis(
        series_name="yB",
        y_axis=[2, 1, 4, 2, 5, 3, 1],
        yaxis_index=1,
        label_opts=opts.LabelOpts(is_show=False),
    )
    .extend_axis(
        yaxis=opts.AxisOpts(
            type_="value",
            name="ybaxis",
            min_='dataMin',
            position="right",
            axistick_opts=opts.AxisTickOpts(is_show=True),
            splitline_opts=opts.SplitLineOpts(is_show=True),    
            #axislabel_opts=opts.LabelOpts(formatter="{value} C"),
            axisline_opts=opts.AxisLineOpts(
                linestyle_opts=opts.LineStyleOpts(color="#5793f3")
            ),            
        )
    )
)
```

```python
line.load_javascript()
```

```python
line.render_notebook()
```
